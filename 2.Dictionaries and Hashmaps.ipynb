Sherlock and Anagrams

```python
import random
import re
import sys
from collections import Counter
# Complete the sherlockAndAnagrams function below.

# def anagram(s,x):
    # for y in s
    # count_s, count_rest = s[0:x], s[x:len(s)]
    # print(f'{count_s} {count_rest}')
    # =)
    # count_rest=Counter(s[x:len(s)])
    # if count_rest-count_s=={}:
    # count_ana+=1
def sherlockAndAnagrams(s):
    count_ana=0
    for x in range(1,len(s)+1):
        a=["".join(sorted(s[y:y+x])) for y in range(len(s)+1-x)]
        b=Counter(a)
        for z in b:
            # Permutacion
            count_ana+=b[z]*(b[z]-1)/2
    return(int(count_ana))

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    q = int(input())

    for q_itr in range(q):
        s = input()

        result = sherlockAndAnagrams(s)

        fptr.write(str(result) + '\n')

    fptr.close()
```
